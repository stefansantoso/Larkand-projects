```python
import pandas as pd
import time
from browsermobproxy import Server
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument("--headless")

server = Server("C:/Users/vin_s/2_projects/Selenium/browsermob-proxy-2.1.4-bin/browsermob-proxy-2.1.4/bin/browsermob-proxy.bat")
server.start()
proxy = server.create_proxy()

proxy.new_har("google")
```

```python
### GET TO MODULE MAPPING PAGE ###

driver = webdriver.Chrome(r'chromedriver.exe', options=chrome_options)
start_url = "https://myedurec.nus.edu.sg/psp/cs90prd/?cmd=login"
driver.get(start_url)
driver.find_element_by_xpath('//*[@id="login"]/div/div[2]/div[2]/span[1]/a').click()
driver.find_element_by_name('UserName').send_keys('[YOUR_USERNAME_HERE]')
driver.find_element_by_name('Password').send_keys('[YOUR_PASSWORD_HERE]')
driver.find_element_by_class_name('submit').click()
time.sleep(2)
driver.find_element_by_xpath('//*[@id="N_STDACAD_SHORTCUT"]').click()
driver.find_element_by_xpath('//*[@id="win0groupletPTNUI_LAND_REC_GROUPLET$11"]').click()
driver.find_element_by_xpath('//*[@id="PT_SIDE$PIMG"]/img').click()

##### Retry until can click panel button ######
def click_banner():
  try:
    driver.find_element_by_xpath('//*[@id="PT_SIDE$PIMG"]').click()
    driver.find_element_by_id('SCC_LO_FL_WRK_SCC_VIEW_BTN$1').click()
  except:
    click_banner()

click_banner()

### CHANGE FRAME ###
driver.switch_to_frame('main_target_win2')
```

```python
def resolve_imaginary(driver):
  driver.refresh()
  time.sleep(1.5)
  click_banner()
  driver.switch_to_frame('main_target_win2')
```

```python
def resolve_processing(driver):
  time.sleep(2)
  if driver.find_element_by_xpath('//*[@id="Delayed_win2"]').value_of_css_property('visibility') == 'hidden':
    print('finish loading')
  else:
    driver.save_screenshot('test.png')
    time.sleep(1)
    resolve_processing(driver)
```
